```python
from software.proto.messages_robocup_ssl_wrapper_pb2 import SSL_WrapperPacket
from python_tools.proto_log import ProtoLog

wrapper_proto_log = ProtoLog("/home/jordan/log_ssl_wrapper_filtered_world_test/SensorFusion_SSL_WrapperPacket", 
                             SSL_WrapperPacket)
```

```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
import numpy as np

output_notebook()

fig = figure(plot_width=900, plot_height=500)
field_lines_proto = wrapper_proto_log[0].geometry.field.field_lines

field_line_xs = []
field_line_ys = []

for fl_proto in field_lines_proto:
    field_line_xs.append([fl_proto.p1.x, fl_proto.p2.x])
    field_line_ys.append([fl_proto.p1.y, fl_proto.p2.y])
    
    
# scale from mm to m
field_line_xs = np.asarray(field_line_xs) / 1000
field_line_ys = np.asarray(field_line_ys) / 1000

fig.multi_line(xs=field_line_xs.tolist(), ys=field_line_ys.tolist())

ROBOT_MAX_RADIUS = 0.09

# plot yellow robots
yellow_robots_proto = wrapper_proto_log[0].detection.robots_yellow
yellow_robot_xs = []
yellow_robot_ys = []

for robot_proto in yellow_robots_proto:
    yellow_robot_xs.append(robot_proto.x)
    yellow_robot_ys.append(robot_proto.y)
    
yellow_robot_xs = np.asarray(yellow_robot_xs) / 1000
yellow_robot_ys = np.asarray(yellow_robot_ys) / 1000

fig.circle(yellow_robot_xs, yellow_robot_ys, radius=ROBOT_MAX_RADIUS, fill_color='yellow', line_color='black')

# plot blue robots
blue_robots_proto = wrapper_proto_log[0].detection.robots_blue
blue_robot_xs = []
blue_robot_ys = []

for robot_proto in blue_robots_proto:
    blue_robot_xs.append(robot_proto.x)
    blue_robot_ys.append(robot_proto.y)
    
blue_robot_xs = np.asarray(blue_robot_xs) / 1000
blue_robot_ys = np.asarray(blue_robot_ys) / 1000

fig.circle(blue_robot_xs, blue_robot_ys, radius=ROBOT_MAX_RADIUS, fill_color='blue', line_color='black')

ball_xs = np.asarray([ball.x for ball in wrapper_proto_log[0].detection.balls]) / 1000
ball_ys = np.asarray([ball.y for ball in wrapper_proto_log[0].detection.balls]) / 1000

fig.circle(ball_xs, ball_ys, radius=0.02, fill_color='orange', line_color='black')

show(fig)
```
